```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('Practise').getOrCreate()
```

```python
df=spark.read.csv('data/narcism.csv',header=True,inferSchema=True)
```

```python
# drop the columns that are not required
df.drop('id').show()
```

```python
df.na.drop().show()
```

```python
##Â any == how
df.na.drop(how='any').show()
```

```python
## treshhold
df.na.drop(thresh=2).show()
```

```python
## subset
df.na.drop(subset=['experience']).show()
```

```python
## fill the missing values
df.na.fill('Missing ID','id').show()
```

```python
df.show()
```

```python
from pyspark.ml.feature import Imputer

imputer = Imputer(
    inputCols=['height','experience'],
    outputCols=['{}_imputed'.format(c) for c in ['height','experience']]
).setStrategy('mean')
```

```python
imputer.fit(df).transform(df).show()
```
