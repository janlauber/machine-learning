```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('ml-basics').getOrCreate()
```

```python
spark
```

```python
df=spark.read.csv('data/narcism.csv',inferSchema=True,header=True)
df.show()
```

```python
## who people over 40
df.filter(df['experience'] > 40).show()
```

```python
df.filter("height > 170").show()
```

```python
df.filter(df['height'] > 170).select(['name','height']).show()
```

```python
# filter people with height > 170 and experience > 40 and are narcissists
df.filter((df['height'] > 170) & (df['experience'] > 40) & (df['is_narcistic'] == True)).show()
```
